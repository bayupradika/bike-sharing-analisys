# Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Wragling
Dataset yang digunakan adalah day.csv, yang berisi 16 kolom dengan data harian mengenai penggunaan sepeda, cuaca, suhu, dan lain-lain.

# Gathering Data

```python
# Load dataset
dday = pd.read_csv('./data/day.csv')
```

# Insight
- Dataset berisi 731 baris dan 16 kolom.
- Data mencakup variabel seperti suhu, kelembaban, dan jumlah penggunaan sepeda.

# Assesing Data

**Insight**
- Tidak ada nilai yang hilang (missing) pada dataset.
- Variabel suhu (temp) memiliki rentang dari sekitar 0.06 hingga 0.86 (sudah dinormalisasi).
- Variabel target, yaitu jumlah total sepeda (cnt), memiliki rentang dari 22 hingga 8714.
- Data musim (season) dikodekan dari 1 hingga 4.

# Cleaning Data

**Insight:**
- Tidak ada nilai kosong atau outlier yang signifikan berdasarkan informasi yang diberikan. Tidak diperlukan proses pembersihan lebih lanjut saat ini.

## Exploratory Data Analysis (EDA)

### Pertanyaan 1: Bagaimana Pengaruh Suhu Terhadap Jumlah Penggunaan Sepeda?
- Visualisasi scatter plot dapat membantu memahami korelasi antara suhu dan jumlah sepeda yang digunakan.

```python
# Visualisasi Pertanyaan 1: Hubungan Suhu dan Jumlah Sepeda
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dday, x='temp', y='cnt')
plt.title('Hubungan Suhu Terhadap Penggunaan Sepeda')
plt.xlabel('Suhu')
plt.ylabel('Jumlah Sepeda')
plt.show()
```

**Insight:**
- Dari plot tersebut, kamu akan melihat apakah ada hubungan positif antara suhu dan jumlah penggunaan sepeda. Semakin tinggi suhu, apakah penggunaan sepeda meningkat atau tidak?

### Pertanyaan 2: Apakah Terdapat Pola Musiman dalam Penggunaan Sepeda?
- Bar plot digunakan untuk melihat apakah ada perbedaan penggunaan sepeda pada musim yang berbeda.

```python
# Visualisasi Pertanyaan 2: Penggunaan Sepeda Berdasarkan Musim
plt.figure(figsize=(10, 6))
sns.barplot(data=dday, x='season', y='cnt')
plt.title('Penggunaan Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Sepeda')
plt.show()
```

**Insight:**
- Dari plot ini, dapat dilihat apakah ada musim tertentu yang lebih ramai digunakan untuk sepeda.
- Setiap musim diwakili oleh angka (1: musim semi, 2: musim panas, 3: musim gugur, 4: musim dingin).

## Visualization & Explanatory Analysis

- Pertanyaan 1: Setelah melihat scatter plot, apakah ada hubungan yang jelas antara suhu dan jumlah penggunaan sepeda? Misalnya, apakah penggunaan sepeda cenderung meningkat ketika suhu naik?
- Pertanyaan 2: Dari bar plot, apa pola musiman yang dapat diidentifikasi? Misalnya, apakah penggunaan sepeda meningkat di musim panas dan menurun di musim dingin?

## Analisis Lanjutan (Opsional
- Analisis clustering menggunakan binning untuk mengkategorikan penggunaan sepeda berdasarkan suhu dan musim.

- RFM Analysis: Analisis perilaku pengguna sepeda berdasarkan frekuensi penggunaan, jumlah penggunaan, dan lainnya.
- Geospatial Analysis: Analisis penggunaan sepeda berdasarkan lokasi geografis jika data lokasi tersedia.
- Clustering: Mengelompokkan pengguna sepeda berdasarkan pola penggunaan tanpa menggunakan algoritma machine learning.

```python
# Binning suhu ke dalam kategori
dday['temp_bin'] = pd.cut(dday['temp'], bins=4, labels=['Rendah', 'Sedang', 'Tinggi', 'Sangat Tinggi'])

# Mengelompokkan data berdasarkan musim dan bin suhu
grouped = dday.groupby(['season', 'temp_bin'], observed=True)['cnt'].mean().reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(data=grouped, x='season', y='cnt', hue='temp_bin', palette='coolwarm')
plt.title('Rata-rata Penggunaan Sepeda Berdasarkan Musim dan Suhu')
plt.xlabel('Musim')
plt.ylabel('Rata-rata Penggunaan Sepeda')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Musim Semi', 'Musim Panas', 'Musim Gugur', 'Musim Dingin'])
plt.legend(title='Kategori Suhu')
plt.show()
```

## Conclusion

1. Conclution pertanyaan 1: Apakah ada hubungan antara suhu dan jumlah penggunaan sepeda?
   - Dari scatter plot, terlihat adanya korelasi positif antara suhu dan jumlah penggunaan sepeda. Penggunaan sepeda cenderung meningkat seiring dengan meningkatnya suhu.
3. Conclution pertanyaan 2: Apakah ada pola musiman yang jelas dalam penggunaan sepeda?
   - Dari bar plot, terlihat bahwa penggunaan sepeda cenderung lebih tinggi di musim panas dan musim semi dibandingkan dengan musim gugur dan musim dingin.
